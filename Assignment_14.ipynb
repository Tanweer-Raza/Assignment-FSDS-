### Q.1. What does RGBA stand for? 

    Ans.1) RGBA stands for Red-Green-Blue-Alpha.
    
           RGBA value is a group of numbers that specify the amount of red, green, blue, and alpha (or 
           transparency) in a color. Each of these component values is an integer from 0 (none at all) to 255 
           (the maximum). These RGBA values are assigned to individual pixels; a pixel is the smallest dot of a 
           single color the computer screen can show.
           e.g Opaque Red is represented as (255, 0, 0, 255)




### Q.2. From the Pillow module, how do you get the RGBA value of any images?

    Ans.2) Pillow is a third-party Python module for interacting with image files. The module has several functions that make it easy to crop, resize, and edit the content of an image. 
    
    Pillow offers the ImageColor.getcolor() function.
    
    
    This function takes a color name string as its first argument, and the string 'RGBA' as its second 
    argument, and it returns an RGBA tuple.
 

    e.g
    from PIL import ImageColor
    ImageColor.getcolor('red', 'RGBA')
    
    Output : (255, 0, 0, 255)

### Q.3. What is a box tuple, and how does it work?

    Ans.3) Box tuple is type of tuple which takes four integer coordinates that represent a rectangular region 
    in an image.The four integers are, in order, as follows:
    Lets say box tuple be represented as box tuple (a,b,c,d)
    
    a : The x-coordinate of the leftmost edge of the box.

    b : The y-coordinate of the top edge of the box.

    c : The x-coordinate of one pixel to the right of the rightmost edge of the box. This integer must be 
    greater than the left integer.

    d : The y-coordinate of one pixel lower than the bottom edge of the box. This integer must be greater 
    than the top integer.
